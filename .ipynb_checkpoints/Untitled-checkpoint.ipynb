```python
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

dfmatch = pd.DataFrame(data[0]['matchSummary'])
dfmatch.head()
```

```python
dfmatch.shape
```

```python
dfmatch.rename({'scorecard':'match_id'},axis=1,inplace= True)
```

```python
dfmatch.head()
```

```python
dfmatch.head()
```

```python
match_id_dict={}

for index, row in dfmatch.iterrows():
   key1=row['team1']+ ' Vs ' + row['team2']
   key2=row['team2']+ ' Vs ' + row['team1']
    
   match_id_dict[key1]=row["match_id"]
   match_id_dict[key2]=row["match_id"]

match_id_dict


    
    
```

"Batting summary"

```python
with open('t20_json_files/t20_wc_batting_summary.json') as g:
    data=json.load(g)
    allrecord=[]

    for r in data:
        allrecord.extend(r['battingSummary'])

```

```python
batdf=pd.DataFrame(allrecord)
batdf.shape
batdf.head()
```

```python
batdf["Out/Not Out"]=batdf.dismissal.apply(lambda x:"out" if len(x)>0 else "Not Out")
batdf.head()
```

```python
batdf.head(10)
```

```python
batdf["match_id"]=batdf.match.map(match_id_dict)
```

```python
batdf.head(15)
```

```python
batdf['batsmanName'] = batdf['batsmanName'].apply(lambda x: x.replace('â€', ''))
batdf['batsmanName'] = batdf['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
batdf.head()
```

```python
  batdf.to_csv('t20_csv_files/battingsummary', index=False)
```

Bowling summary


```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_id_dict)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

Process Player Information

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv('t20_csv_files/dim_players_no_images.csv', index = False)
```
